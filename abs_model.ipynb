```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn import tree
import pandas as pd
```

```python
df = pd.read_csv('UpdatedEnrollmentDataFrame.csv')
```

```python
df.describe()
```

```python
df['Application Created Date'] = pd.to_datetime(df['Application Created Date'])
df['Application Created Month'] = df['Application Created Date'].dt.month
df
```

```python
df_decided = df[df['Application Status'] == 'Decided']
```

```python
df_decided.describe(include='all')
```

```python
decision_reasons = [
    'Deposit - Not Required',
    'Admit Declined',
    'Admit Declined - Post Enroll(CF)',
    'Update Major - Post Decision',
    'Waitlist Invite',
    'Update Term - Post Decision',
    'Admit Regular',
    'Update Term & Major - Post Decision',
    'Admit Conditional',
    'Update Term - Post Enroll(CF)',
    'Waitlist Accepted',
    'Rejected - Waitlist Deny',
    'Decision Delayed',
    'Wailist Declined',
    'Deposit Paid',
    'No Show - Post Enroll(CF)',
    'Admit Probation',
    'Admit Conditional & Probationary',
    'Update Term & Major - Post Enroll(CF)',
    'Update Major - Post Enroll(CF)',
    'Deposit Waived/Not Required',
    'Admit Conditional & Bridge'
]
df_matriculate = df_decided[df_decided['Decision Reason'].isin(decision_reasons)]
df_matriculate['Decision Reason'].describe()
```

```python
df_matriculate['Matriculate'] = 0
df_matriculate.loc[df_matriculate['Applications Registered in Colleague'] == 0, 'Matriculate'] = 0
df_matriculate.loc[df_matriculate['Applications Registered in Colleague'] == 1, 'Matriculate'] = 1
```

```python
df_matriculate['GMAT'] = df_matriculate['Tests - GMAT Total'].notna().astype(int)
df_matriculate['GMAT'].describe()
```

```python
selected_columns = ['Application Referral Source', 'Application Program', 'Application Start Term', 
                    'Application Created Month',
                    'Decision Reason', 'Country', 'State', 'Visa Type', 'Age', 'Race', 'Sex', 
                    'Intend to apply for financial aid', 'GPA (Converted) Category', 
                    'Scholarship Category', 'GMAT', 'Matriculate']

df_model = df_matriculate.loc[:, selected_columns]
```

```python
#WA:Seeing if applicant is from WA
df_model['WA'] = 0
df_model.loc[df_model['State'] == 'WA', 'WA'] = 1
```

```python
#Country Category:
df_model['Country Category'] = 'Other Countries'
df_model.loc[df_model['Country'] == 'United States', 'Country Category'] = 'United States'
df_model.loc[df_model['Country'] == 'India', 'Country Category'] = 'India'
df_model.loc[df_model['Country'] == 'China', 'Country Category'] = 'China'
df_model.loc[df_model['Country'] == 'Nigeria', 'Country Category'] = 'Nigeria'
df_model
```

```python
#Decision Tree for Matriculation
#Key finding1 for this Decision Tree: Applicants From WA are more likely to Matriculate
#Key finding2 for this Decision Tree: Applicants with Not Valid for ScholarShip are less likely to Matriculate
#Key finding3 for this Decision Tree: Applicants that don't have referral source are less likely to Matriculate
#Key finding4 for this Decision Tree: Applicants that are under 25(just graduated from university) are likely to Matriculate


X = df_model[['Application Referral Source',
 'Application Program',
 'Visa Type',
 'Age',
 'Race',
 'Sex',
 'Intend to apply for financial aid',
 'GPA (Converted) Category',
 'Scholarship Category',
 'GMAT',
 'Country Category',
 'WA']]
X_encoded = pd.get_dummies(X)
y = df_model['Matriculate']
clf = tree.DecisionTreeClassifier(max_depth=5,min_samples_leaf=10)
clf = clf.fit(X_encoded, y)
tree.export_graphviz(clf, out_file='matriculateDtree.dot',
feature_names=X_encoded.columns.tolist(), class_names=[str(cls) for cls in clf.classes_], filled=True)
```

```python
#Decision Tree for Matriculation(Non-American)
#Key finding1 for this Decision Tree: Applicants who Matriculate their Scholarship Category are all non not Not valid for Award

df_model['American'] = 0
df_model.loc[df_model['Country'] == 'United States', 'American'] = 1
df_notUS = df_model[df_model['American'] != 1]
X = df_notUS[['Application Referral Source',
 'Application Program',
 'Visa Type',
 'Age',
 'Race',
 'Sex',
 'Intend to apply for financial aid',
 'GPA (Converted) Category',
 'Scholarship Category',
 'GMAT',
 'Country Category',
 'WA']]
X_encoded = pd.get_dummies(X)
y = df_notUS['Matriculate']
clf = tree.DecisionTreeClassifier(max_depth=5,min_samples_leaf=10)
clf = clf.fit(X_encoded, y)
tree.export_graphviz(clf, out_file='NotUSmatriculateDtree.dot',
feature_names=X_encoded.columns.tolist(), class_names=[str(cls) for cls in clf.classes_], filled=True)
```

```python
#Decision Tree for Declined

df_model['Declined'] = 0
df_model.loc[df_model['Decision Reason'] == 'Admit Declined', 'Declined'] = 1
df_model.loc[df_model['Decision Reason'] == 'Admit Declined - Post Enroll(CF) ', 'Declined'] = 1
X = df_model[['Application Referral Source',
 'Application Program',
 'Visa Type',
 'Age',
 'Race',
 'Sex',
 'Intend to apply for financial aid',
 'GPA (Converted) Category',
 'Scholarship Category',
 'GMAT',
 'Country Category',
 'WA']]
X_encoded = pd.get_dummies(X)
y = df_model['Declined']
clf = tree.DecisionTreeClassifier(max_depth=3,min_samples_leaf=10)
clf = clf.fit(X_encoded, y)
tree.export_graphviz(clf, out_file='DeclinedDtree.dot',
feature_names=X_encoded.columns.tolist(), class_names=[str(cls) for cls in clf.classes_], filled=True)

```

```python
#Decision Tree for data removing Desion Reason: Admit Declined & Admit Declined - Post Enroll(CF)
df_without_declined = df_model[df_model['Declined'] != 1]
X = df_without_declined[['Application Referral Source',
 'Application Program',
 'Visa Type',
 'Age',
 'Race',
 'Sex',
 'Intend to apply for financial aid',
 'GPA (Converted) Category',
 'Scholarship Category',
 'GMAT',
 'Country Category',
 'WA']]
X_encoded = pd.get_dummies(X)
y = df_without_declined['Matriculate']
clf = tree.DecisionTreeClassifier(max_depth=3,min_samples_leaf=10)
clf = clf.fit(X_encoded, y)
tree.export_graphviz(clf, out_file='WithoutDeclinedDtree.dot',
feature_names=X_encoded.columns.tolist(), class_names=[str(cls) for cls in clf.classes_], filled=True)
```

```python
#Decision Tree for MSBA Matriculation
df_model['MSBA'] = 0
df_model.loc[df_model['Application Program'] == 'Business Analytics - MSBA', 'MSBA'] = 1
df_MSBA = df_model[df_model['MSBA'] == 1]
X = df_MSBA[['Application Referral Source',
 'Visa Type',
 'Age',
 'Race',
 'Sex',
 'Intend to apply for financial aid',
 'GPA (Converted) Category',
 'Scholarship Category',
 'GMAT',
 'Country Category',
 'WA']]
X_encoded = pd.get_dummies(X)
y = df_MSBA['Matriculate']
clf = tree.DecisionTreeClassifier(max_depth=3,min_samples_leaf=10)
clf = clf.fit(X_encoded, y)
tree.export_graphviz(clf, out_file='MSBADtree.dot',
feature_names=X_encoded.columns.tolist(), class_names=[str(cls) for cls in clf.classes_], filled=True)
```

```python
#Decision Tree for MSBA Matriculation(Non-American)
df_notUS_MSBA = df_MSBA[df_MSBA['American'] != 1]
X = df_notUS_MSBA[['Application Referral Source',
 'Visa Type',
 'Age',
 'Race',
 'Sex',
 'Intend to apply for financial aid',
 'GPA (Converted) Category',
 'Scholarship Category',
 'GMAT',
 'Country Category',
 'WA']]
X_encoded = pd.get_dummies(X)
y = df_notUS_MSBA['Matriculate']
clf = tree.DecisionTreeClassifier(max_depth=3,min_samples_leaf=10)
clf = clf.fit(X_encoded, y)
tree.export_graphviz(clf, out_file='notUSMSBADtree.dot',
feature_names=X_encoded.columns.tolist(), class_names=[str(cls) for cls in clf.classes_], filled=True)
```
