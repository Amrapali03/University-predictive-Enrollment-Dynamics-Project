## Import

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Read data

```python
# read data
enrollment_data = pd.read_csv('ABS _ Recruiting and Enrollment Data 20240411-161419.csv')
enrollment_data.head(5)
```

## Transformations

```python
# Convert dates to datetime format where applicable
enrollment_data['Record Created Date'] = pd.to_datetime(enrollment_data['Record Created Date'], errors = 'coerce')
enrollment_data['Application Created Date'] = pd.to_datetime(enrollment_data['Application Created Date'], errors = 'coerce')
enrollment_data['Decisions Released Date'] = pd.to_datetime(enrollment_data['Decisions Released Date'], errors = 'coerce')
enrollment_data.head(4)
 

```

```python
enrollment_data = enrollment_data.drop_duplicates()
```

```python
enrollment_data['Application Program'].unique()
```

## What is the 4-year enrollment trend and how has it shifted?
Analyze Enrollment Trends: We will examine enrollment trends over the last four years by program.
Not considering 2024 as part of historical trend analysis

```python

# Filter data for the last four years only
recent_data = enrollment_data[(enrollment_data['Application Created Date'].dt.year >= 2020) & (enrollment_data['Application Created Date'].dt.year <= 2023)]

# Group by year and count the number of applications
enrollment_trend = recent_data.groupby(recent_data['Application Created Date'].dt.year).size()

enrollment_trend

# Plotting the trend
plt.figure(figsize=(10, 5))
enrollment_trend.plot(kind='line', marker='o', linestyle='-', color='b', linewidth=2)
plt.title('4-Year Enrollment Trend based on Applications created')
plt.xlabel('Year')
plt.ylabel('Number of Applications')
plt.grid(True)
plt.xticks(list(range(2020, 2024)))  # Ensure all years are shown
plt.show()
```

```python
# Filter data for the last four years only
recent_data = enrollment_data[(enrollment_data['Decisions Released Date'].dt.year >= 2020) & (enrollment_data['Decisions Released Date'].dt.year <= 2023)]

# Group by year and count the number of applications
enrollment_trend = recent_data.groupby(recent_data['Decisions Released Date'].dt.year).size()

enrollment_trend

# Plotting the trend
plt.figure(figsize=(10, 5))
enrollment_trend.plot(kind='line', marker='o', linestyle='-', color='b', linewidth=2)
plt.title('4-Year Enrollment Trend based on Decision released')
plt.xlabel('Year')
plt.ylabel('Number of Applications')
plt.grid(True)
plt.xticks(list(range(2020, 2024)))  # Ensure all years are shown
plt.show()
```

Compare both the dates

```python
# Ensure both date columns are in datetime format
enrollment_data['Application Created Date'] = pd.to_datetime(enrollment_data['Application Created Date'], errors='coerce')
enrollment_data['Decisions Released Date'] = pd.to_datetime(enrollment_data['Decisions Released Date'], errors='coerce')

# Filter data for the last four years only
recent_data = enrollment_data[(enrollment_data['Application Created Date'].dt.year >= 2020) & (enrollment_data['Application Created Date'].dt.year <= 2023)]

# Group by year and count the number of applications
application_trend = recent_data.groupby(recent_data['Application Created Date'].dt.year).size()

# Similarly, group by year for the decisions released
decision_trend = recent_data.groupby(recent_data['Decisions Released Date'].dt.year).size()

# Plotting both trends
plt.figure(figsize=(12, 6))
application_trend.plot(kind='line', marker='o', linestyle='-', color='b', linewidth=2, label='Application Created')
decision_trend.plot(kind='line', marker='x', linestyle='-', color='r', linewidth=2, label='Decision Released')
plt.title('4-Year Application Trends Comparison')
plt.xlabel('Year')
plt.ylabel('Number of Applications / Decisions')
plt.legend()
plt.grid(True)
plt.xticks(list(range(2020, 2024)))  # Ensure all years are shown
plt.show()
```

## Trend by Programs

```python
# Convert 'Application Created Date' to datetime and clean data
enrollment_data['Application Created Date'] = pd.to_datetime(enrollment_data['Application Created Date'], errors='coerce')
enrollment_data = enrollment_data.dropna(subset=['Application Created Date', 'Application Program'])
enrollment_data = enrollment_data[(enrollment_data['Application Created Date'].dt.year >= 2020) & (enrollment_data['Application Created Date'].dt.year <= 2023)]

# Group by year and application program
grouped_data = enrollment_data.groupby([enrollment_data['Application Created Date'].dt.year, 'Application Program']).size().unstack(fill_value=0)

# Plotting the data
fig, ax = plt.subplots(figsize=(14, 8))
grouped_data.plot(kind='bar', ax=ax, stacked=True)
plt.title('Enrollment Trend by Application Program (2020-2023)')
plt.xlabel('Year')
plt.ylabel('Number of Applications')
plt.legend(title='Application Program', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=0)
plt.grid(True)
plt.tight_layout()
plt.show()
```

Program categories grouped together
